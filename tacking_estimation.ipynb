## Direction-Only Dependent Finsler Metrics

### Defining Finsler Geometry

```python
class EllipticFinsler(ABC):
    def __init__(self,
                 c1=1.0, 
                 c2=1.0,
                 a=1.0,
                 b=1.0,
                 theta=jnp.pi/4
                )->None:
        
        self.c1 = c1
        self.c2 = c2
        self.a = a
        self.b = b
        self.theta = theta
        
    def __call__(self, t, x, v):
        
        c1 = self.c1
        c2 = self.c2
        a = self.a
        b = self.b
        theta = self.theta
        
        x,y = v[0], v[1]
        
        a2 = a**2
        b2 = b**2
        a4 = a**4
        b4 = b**4
        c12 = c1**2
        c22 = c2**2

        f = (-a2*c2*(x*jnp.sin(theta)+y*jnp.cos(theta))-\
             b2*c1*(x*jnp.cos(theta)-y*jnp.sin(theta))+\
             (a4*b2*(x*jnp.sin(theta)+y*jnp.cos(theta))**2+\
              a2*b4*(x*jnp.cos(theta)-y*jnp.sin(theta))**2-\
              a2*b2*c12*(x*jnp.sin(theta)+y*jnp.cos(theta))**2+\
              2*a2*b2*c1*c2*(x*jnp.cos(theta)-y*jnp.sin(theta))*(x*jnp.sin(theta)+y*jnp.cos(theta))-\
              a2*b2*c22*(x*jnp.cos(theta)-y*jnp.sin(theta))**2)**(1/2))/(a2*b2-a2*c22-b2*c12)
        
        return f
```

```python
rho=3/2
phi=jnp.pi/2+6*jnp.pi/10
theta = jnp.pi
a=2
b=2
c1 = -rho*jnp.cos(phi)
c2 = -rho*jnp.sin(phi)

Falpha = EllipticFinsler(c1=c1,c2=c2, a=a,b=b,theta=theta)
Malpha = LorentzFinslerManifold(F=Falpha)
```

```python
a=1
b=1
c1=3/4
c2=0
theta=0

Fbeta = EllipticFinsler(c1=c1,c2=c2, a=a,b=b,theta=theta)
Mbeta = LorentzFinslerManifold(F=Fbeta)
```

```python
tack_metrics = [Malpha,Mbeta,Malpha,Mbeta,Malpha]
reverse_tack_metrics = [Mbeta, Malpha, Mbeta, Malpha, Mbeta]
```

### Defining Geodesic and Tacking Optimization

```python
Geodesic = GEORCE_H(Malpha, init_fun=None, T=T, tol=tol, max_iter=max_iter, line_search_params={'rho': 0.5})
ReverseGeodesic = GEORCE_H(Mbeta, init_fun=None, T=T, tol=tol, max_iter=max_iter, line_search_params={'rho': 0.5})
```

```python
Tacking = SequentialOptimizationBFGS(tack_metrics, init_fun=None, max_iter=max_iter, 
                                     tol=tol, T=T, sub_iter=sub_iter, line_search_params={'rho': 0.5})
ReverseTacking = SequentialOptimizationBFGS(reverse_tack_metrics, init_fun = None, max_iter=max_iter, tol=tol,
                                            T=T, sub_iter=sub_iter, line_search_params={'rho': 0.5})
```

### Estimating Tack Points

```python
#Estimating initial geodesic
zt1 = Geodesic(t0, z0, zT)
zt2 = ReverseGeodesic(t0, z0, zT)
print(f"Computing geodesics...")
tack_curves = [zt1[1]]
reverse_tack_curves = [zt2[1]]
time_curves = [zt1[0]]
reverse_time_curves = [zt2[0]]
print(f"\t -Tavel time for alpha metric: {zt1[0][-1]:.4f}")
print(f"\t -Tavel time for beta metric: {zt2[0][-1]:.4f}")
for i in range(1,len(tack_metrics)):
    print(f"Estimation {i} tack points...")
    zt1 = jit(lambda t0, z0, zT: Tacking(t0, z0, zT, n_tacks=i))(t0,z0,zT)
    zt2 = jit(lambda t0, z0, zT: ReverseTacking(t0, z0, zT, n_tacks=i))(t0,z0,zT)
    tack_curves.append(zt1[1])
    reverse_tack_curves.append(zt2[1])
    time_curves.append(zt1[0])
    reverse_time_curves.append(zt2[0])
    print(f"\t -Tavel time for tack curve: {zt1[0][-1]:.4f}")
    print(f"\t -Tavel time for reverse tack curve: {zt2[0][-1]:.4f}")
    print(f"\t -Idx for tack curve: {zt1[-1]}")
    print(f"\t -Idx for reverse tack curve: {zt2[-1]}")
    print(f"\t -Grad for tack curve: {jnp.linalg.norm(zt1[-2]):.4f}")
    print(f"\t -Grad for reverse tack curve: {jnp.linalg.norm(zt2[-2]):.4f}")
print("Estimation Completed")
```

### Defining Finsler Geometry

```python
class Falpha(ABC):
    def __init__(self,
                 d:float=0.5,
                )->None:
        
        self.d = d
        
    def __call__(self, t, z, dz):

        u, v = z[0], z[1]
        x,y = dz[0], dz[1]
        
        d = self.d
        d2 = d**2
        
        return (-jnp.sqrt((x**2)+(y**2)-((x-y)**2)*d2)+(x+y)*d)/(2*d2*v-v)
        
        
```

```python
class Fbeta(ABC):
    def __init__(self,
                 d:float=0.5,
                )->None:
        
        self.d = d
        
    def __call__(self, t, z, dz):

        u, v = z[0], z[1]
        x,y = dz[0], dz[1]
        
        d = self.d
        d2 = d**2
        
        return (-jnp.sqrt((x**2)+(y**2)-((x+y)**2)*d2)+(x-y)*d)/(2*d2*v-v)
```

```python
d=0.5

Falpha = Falpha(d=0.5)
Malpha = LorentzFinslerManifold(F=Falpha)
```

```python
d=0.5

Fbeta = Fbeta(d=0.5)
Mbeta = LorentzFinslerManifold(F=Fbeta)
```

### Defining Metric and points

```python
tack_metrics = [Malpha,Mbeta,Malpha,Mbeta,Malpha]
reverse_tack_metrics = [Mbeta, Malpha, Mbeta, Malpha, Mbeta]
```

```python
k = 10.
t0 = jnp.zeros(1, dtype=jnp.float32).squeeze()
z0 = jnp.array([1.,1.], dtype=jnp.float32)
zT = jnp.array([k,1.], dtype=jnp.float32)
```

### Defining Geodesic and Tacking Optimization

```python
Geodesic = GEORCE_H(Malpha, init_fun=None, T=T, tol=tol, max_iter=max_iter, line_search_params={'rho': 0.5})
ReverseGeodesic = GEORCE_H(Mbeta, init_fun=None, T=T, tol=tol, max_iter=max_iter, line_search_params={'rho': 0.5})
```

```python
Tacking = SequentialOptimizationBFGS(tack_metrics, init_fun=None, max_iter=max_iter, 
                                     tol=tol, T=T, sub_iter=sub_iter, line_search_params={'rho': 0.5})
ReverseTacking = SequentialOptimizationBFGS(reverse_tack_metrics, init_fun = None, max_iter=max_iter, tol=tol,
                                            T=T, sub_iter=sub_iter, line_search_params={'rho': 0.5})
```

```python
Tacking = SequentialOptimizationADAM(tack_metrics, lr_rate=0.01, init_fun=None, max_iter=max_iter, 
                                     tol=tol, T=T, sub_iter=sub_iter, line_search_params={'rho': 0.5})
ReverseTacking = SequentialOptimizationADAM(reverse_tack_metrics, lr_rate = 0.01, init_fun = None, max_iter=max_iter, tol=tol,
                                            T=T, sub_iter=sub_iter, line_search_params={'rho': 0.5})
```

### Estimating Tack Points

```python
#Estimating initial geodesic
zt1 = Geodesic(t0, z0, zT)
zt2 = ReverseGeodesic(t0, z0, zT)
print(f"Computing geodesics...")
tack_curves = [zt1[1]]
reverse_tack_curves = [zt2[1]]
time_curves = [zt1[0]]
reverse_time_curves = [zt2[0]]
print(f"\t -Tavel time for alpha metric: {zt1[0][-1]:.4f}")
print(f"\t -Tavel time for beta metric: {zt2[0][-1]:.4f}")
for i in range(1,len(tack_metrics)):
    print(f"Estimation {i} tack points...")
    zt1 = jit(lambda t0, z0, zT: Tacking(t0, z0, zT, n_tacks=i))(t0,z0,zT)
    zt2 = jit(lambda t0, z0, zT: ReverseTacking(t0, z0, zT, n_tacks=i))(t0, z0, zT)
    tack_curves.append(zt1[1])
    reverse_tack_curves.append(zt2[1])
    time_curves.append(zt1[0])
    reverse_time_curves.append(zt2[0])
    print(f"\t -Tavel time for tack curve: {zt1[0][-1]:.4f}")
    print(f"\t -Tavel time for reverse tack curve: {zt2[0][-1]:.4f}")
    print(f"\t -Idx for tack curve: {zt1[-1]}")
    print(f"\t -Idx for reverse tack curve: {zt2[-1]}")
    print(f"\t -Grad for tack curve: {jnp.linalg.norm(zt1[-2]):.4f}")
    print(f"\t -Grad for reverse tack curve: {jnp.linalg.norm(zt2[-2]):.4f}")
print("Estimation Completed")
```
